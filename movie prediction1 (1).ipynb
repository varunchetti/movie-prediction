```python
x1=df.drop(['Lead_Actress_rating','Director_rating','Producer_rating'],axis=1)
```

```python
y1=df1['Collection']
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.25,random_state=2549)
```

```python
lr1=LinearRegression()
```

```python
lr1.fit(x1_train,y1_train)
```

```python
lr1.score(x1_test,y1_test)
```

```python
lr1.score(x1_train,y1_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=70,n_jobs=50,max_depth=3)
```

```python
 rf.fit(x1_train,y1_train)
```

```python
rf.score(x1_test,y1_test)
```

```python
rf.score(x1_train,y1_train)
```

```python
param={
    
         'n_estimators':[10,20,30,40,50],
         'max_depth':[None,5,10,20],
         'n_jobs':[50,70,90,100]
    
}
    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gr=GridSearchCV(estimator=rf,param_grid=param,cv=5)
```
